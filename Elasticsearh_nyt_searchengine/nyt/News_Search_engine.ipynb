```python
import elasticsearch
import json

es = elasticsearch.Elasticsearch()

with open('test.json', 'r') as fp:
    articles = json.load(fp)

for i, a in enumerate(articles[:500]):
    es.index(index='scrappy', body = a, id = i, doc_type='article')
```

```python
q1 = es.search(index='scrappy', q = 'body:"Francis"')
q1
```

```python
q2 = es.search(index='scrappy', q = 'title:"Starbucks"')
q2
```

```python
q3 = es.search(index='scrappy', q = 'authors:"Timothy Egan"')
q3
```

```python
q4_1 = es.search(index='scrappy', q = 'austrio')
q4_2 = es.search(index='scrappy', q = 'austria')

print(q4_1['hits']['total'])
print(q4_2['hits']['total'])
```

```python
q5 = es.search(index='scrappy', q = 'authors:"Timothy Egan" body:"American"')
q5
```
